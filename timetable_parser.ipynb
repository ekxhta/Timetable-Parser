```python

# Load all sheets from the Excel file into a dictionary of DataFrames
dfs = pd.read_excel('ODDSEM2024.xls', sheet_name=None)

# Ask the user to input the semester number
semester = input("Enter the semester number: ")

# Access the specific sheet corresponding to the entered semester
sheet_name = f"BTECH {semester} SEM"
if sheet_name in dfs:
    df_semester = dfs[sheet_name]
else:
    print(f"Sheet for semester {semester} not found!")
    exit()

# Assuming df_semester is a DataFrame from pandas and is already defined
day_to_classes = {}
day_indices = [1, 17, 33, 49, 65, 81]
for i in range(len(day_indices)):
  day_index=day_indices[i]
  day_name =df_semester.iloc[day_index,0]
  # print(day_name, 'is being parsed')


  for j in range(1,9):
    time=df_semester.iloc[0,j]
    # print(f"Classes at :  {time}")

    class_list=[]

    for k in range(0,16):
      class_name=df_semester.iloc[day_indices[i]+k,j]
      if pd.notna(class_name):
        class_list.append(class_name)

    if day_name not in day_to_classes:
          day_to_classes[day_name] = []

    # Map the day to a tuple of time and list of class names
    day_to_classes[day_name].append ({
        "time": time,
        "classes": class_list
    })  # Store time and classes as a dictionary

    # Display the day_to_classes dictionary on separate lines
for day, class_info in day_to_classes.items():
    print(f"Day: {day}")
    for info in class_info:
        print(f"Time: {info['time']}")
        print("Classes:")
        for class_name in info['classes']:
            print(f" - {class_name}")  # Indent the class names for clarity
        print()  #
```

**To obtain timetable of all batches according to the day of the week**

```python
def timetable(day_needed):
  for day, class_info in day_to_classes.items():
    if day==day_needed:
      print(f"Day: {day}")
      for info in class_info:
          print(f"Time: {info['time']}")
          print("Classes:")
          for class_name in info['classes']:
              print(f" - {class_name}")  # Indent the class names for clarity
          print()
```

**To obtain timetable based on Batch of the student and Day of the week concerned :**
